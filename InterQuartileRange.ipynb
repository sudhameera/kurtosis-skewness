```python
def FreqTable (columnName,data):
    freqTable=pd.DataFrame(columns=['Unique_values','frequency','Rfreq','Cumsum'])
    freqTable['Unique_values']=data[columnName].value_counts().index
    freqTable['frequency']=data[columnName].value_counts().values
    freqTable['Rfreq']=freqTable['frequency']/103
    freqTable['Cumsum']=freqTable['Rfreq'].cumsum()
    return freqTable
```

# skewness , kurtosis

```python
data['hsc_p'].skew()
```

```python
def univariate(quan,data):
    preprocessed=pd.DataFrame ( index=['mean','meadian','mode','Q1:25%','Q2:50%','Q3:75%','99%',
                                'Q4:100%','IQR','1.5rule','Lesser','Greater','Min',
                                       'Max','skewness','kurtosis'],columns=quan)
    for columnName in quan:
        preprocessed[columnName]['mean']=data[columnName].mean()
        preprocessed[columnName]['meadian']=data[columnName].median()
        preprocessed[columnName]['mode']=data[columnName].mode()[0]
        preprocessed[columnName]['99%']=np.percentile(data[columnName],99)
        preprocessed[columnName]['Q1:25%']=data.describe()[columnName]['25%']
        preprocessed[columnName]['Q2:50%']=data.describe()[columnName]['50%']
        preprocessed[columnName]['Q3:75%']=data.describe()[columnName]['75%']
        preprocessed[columnName]['Q4:100%']=data.describe()[columnName]['max']
        preprocessed[columnName]['IQR']= preprocessed[columnName]['Q3:75%']-preprocessed[columnName]['Q1:25%']
        preprocessed[columnName]['1.5rule']=1.5*preprocessed[columnName]['IQR']
        preprocessed[columnName]['Lesser']= preprocessed[columnName]['Q1:25%']-preprocessed[columnName]['1.5rule']
        preprocessed[columnName]['Greater']= preprocessed[columnName]['Q3:75%']+ preprocessed[columnName]['1.5rule']
        preprocessed[columnName]['Min'] = data[columnName].min()
        preprocessed[columnName]['Max'] = data[columnName].max()
        preprocessed[columnName]['skewness']=data[columnName].skew()
        preprocessed[columnName]['kurtosis']=data[columnName].kurtosis()
        
    return preprocessed
```

```python
univariate(quan,data)
```
